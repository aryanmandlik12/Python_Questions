### 14. Given a string s representing time in 24-hour format "HH:MM", compute the smallest angle in degrees between the hour and minute hands of an analog clock.

```python
time = input("Enter the time in HH:MM format separeted by colon:")
HH = int(time[0:2])
HH = HH%12
MM = int(time[3:])
minute_angle = MM * 6
hour_angle = 30 * HH + 0.5 * MM

angle = abs(hour_angle - minute_angle)
minimum_angle = min(angle, 360 - angle)

print("The minimum angle between the hour hand and minute hand is", minimum_angle)
```

### 15. Given two rectangles, find if the given two rectangles overlap or not. A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle). 

```python
L1 = input("Enter the coordinates of the top left corner of first rectanle separeted by comma:")
R1 = input("Enter the coordinates of the top bottom right of first rectanle separeted by comma:")
L2 = input("Enter the coordinates of the top left corner of second rectanle separeted by comma:")
R2 = input("Enter the coordinates of the top bottom right of second rectanle separeted by comma:")

L1_parts = L1.split(",")
R1_parts = R1.split(",")
L2_parts = L2.split(",")
R2_parts = R2.split(",")

L1_x = float(L1_parts[0])
L1_y = float(L1_parts[1])
R1_x = float(R1_parts[0])
R1_y = float(R1_parts[1])
L2_x = float(L2_parts[0])
L2_y = float(L2_parts[1])
R2_x = float(R2_parts[0])
R2_y = float(R2_parts[1])

if R1_x < L2_x or R2_x < L1_x:
    print("The Rectangles does not overlap")
elif R1_y > L2_y or R2_y > L1_y:
    print("The Rectangle does not overlap")
else:
    print("The rectangles overlap")
```

### 28. Write a program to print whether a given number is prime number or not

```python
n = int(input("Enter the number you want to check"))

if n <= 2:
        print("The given number is not a prime")
else:
    is_prime = True

    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print("The given number is a prime number")
    else:
        print("The given number is not a prime number")
```

### 32.  User will provide 2 numbers you have to find the HCF of those 2 numbers

```python
a = int(input("Enter first number:"))
b = int(input("Enter second number:"))
if a<b:
    n = a
else: 
    n = b
list = []
for i in range (1, n+1):
    if a%i == 0 and b%i == 0:
        list.append(i)
HCF = max(list)
print("The hcf of", a, "and", b, "is", HCF)  
```

### 59. Write a python program to convert a string to title case without using the title()

```python
s = input("Enter a string: ")

result = ""
new_word = True

for ch in s:
    if ch == " ":
        result += ch
        new_word = True
    else:
        if new_word:
            result += ch.upper()
            new_word = False
        else:
            result += ch.lower()

print(result)
```

### 66. Write a program to check if a list is in ascending order or not

### 69. Write a program to replace an item with a different item if found in the list

```python
l=list(map(int, input("Enter all objects of list separated by space: ").split())) 
s = int(input("Enter the number which you want to remove"))

if s in l:
    m = int(input("Enter the number which you want to add: "))
    for i in range(len(l)):      # iterate by index, not value
        if l[i] == s:
            l[i] = m             # direct assignment, no pop/insert needed
else:
    print(f"{s} not found in list")

print("Modified list:", l)
```

### 70. Write a program that can convert a 2D list to 1D list. Write a program that can print

### 77. Write a program that can sort a given unsorted list. Dont use any built in  function for sorting.

```python
l = list(map(int, input("Enter all objects of unsorted list separated by space: ").split()))

n = len(l)

for i in range(n):
    for j in range(0, n - i - 1):   # -i because last i elements are already sorted
        if l[j] > l[j+1]:
            l[j], l[j+1] = l[j+1], l[j]   # swap

print("Sorted list:", l)
```

### 80. Write a program to merge two given dictionary

```python
d1 = {}
n1 = int(input("How many key-value pairs for 1st dict: "))

for i in range(n1):
    key   = input(f"Enter key {i+1}: ")
    value = input(f"Enter value {i+1}: ")
    d1[key] = value


d2 = {}
n2 = int(input("How many key-value pairs for 2nd dict: "))

for i in range(n1):
    key   = input(f"Enter key {i+1}: ")
    value = input(f"Enter value {i+1}: ")
    d2[key] = value

merged ={}

for key in d1:
    merged[key] = d1[key] 

for key in d2:
    merged[key] = d2[key] 

print("Merged dictionary:", merged)
```

### 81. Write a program to swap the key value pair for max and min values
### Eg if the dict is like this {‘a’:1, ’b’:2, ’c’:3}
### Output should be {a:3,b:2,c:1}

```python
d1 = {}
n1 = int(input("How many key-value pairs for 1st dict: "))

for i in range(n1):
    key   = input(f"Enter key {i+1}: ")
    value = input(f"Enter value {i+1}: ")
    d1[key] = value

values = list(d1.values())

max_val = max(values)
min_val = min(values)

max_key = [key for key, value in d1.items() if value == max_val]
min_key = [key for key, value in d1.items() if value == min_val]

d1[min_key[0]] = max_val
d1[max_key[0]] = min_val

print("Edited Dictionary", d1)
```

### 82. Write a program to find histogram of a given set of numbers. Take bin size from user. Print the result in the form of a dictionary.

```python
numbers  = list(map(int, input("Enter numbers separated by space: ").split()))
bin_size = int(input("Enter bin size: "))

min_val  = min(numbers)
max_val  = max(numbers)

histogram = {}

start = min_val
while start <= max_val:
    end   = start + bin_size - 1 

    count = 0
    for num in numbers:          
        if start <= num <= end:
            count += 1

    key = f"{start}-{end}"  
    histogram[key] = count           

    start += bin_size                  

print("Histogram:", histogram)
```

#### 84. Write a function that accepts a list of strings and performs Bag of words and convert it to numerical vectors.
https://en.wikipedia.org/wiki/Bag-of-words_model

```python
def bag_of_words(sentences):

    # ---- Step 1: Build Vocabulary ----
    vocabulary = []

    for sentence in sentences:
        words = sentence.lower().split()    
        for word in words:
            if word not in vocabulary:       
                vocabulary.append(word)

    vocabulary.sort()      
    print("Vocabulary:", vocabulary)
    print("Vocab size:", len(vocabulary))

    # ---- Step 2: Build Vectors ----
    vectors = []

    for sentence in sentences:
        words = sentence.lower().split()

        vector = []
        for vocab_word in vocabulary:
            count = 0
            for word in words:
                if word == vocab_word:
                    count += 1             
            vector.append(count)

        vectors.append(vector)

    result = {}
    for i in range(len(sentences)):
        result[sentences[i]] = vectors[i]

    return result

n = int(input("How many sentences: "))
sentences = []
for i in range(n):
    s = input(f"Enter sentence {i+1}: ")
    sentences.append(s)

output = bag_of_words(sentences)

print("\n--- Bag of Words Vectors ---")
for sentence, vector in output.items():
    print(f"{sentence} → {vector}")
```

### 85. Write a dummy program that can perform login and registration using a menu driven program

```python
users = {}

def register():
    print("\n--- Register ---")
    username = input("Enter username: ")

    if username in users:
        print("Username already exists. Try a different one.")
        return

    password = input("Enter password: ")
    confirm  = input("Confirm password: ")

    if password != confirm:
        print("Passwords do not match. Try again.")
        return

    users[username] = password
    print(f"Registration successful! Welcome {username}.")


def login():
    print("\n--- Login ---")
    username = input("Enter username: ")

    if username not in users:
        print("Username not found. Please register first.")
        return

    password = input("Enter password: ")

    if users[username] == password:
        print(f"Login successful! Welcome back, {username}.")
    else:
        print("Incorrect password. Try again.")


def show_menu():
    print("\n======= MENU =======")
    print("1. Register")
    print("2. Login")
    print("3. Exit")
    print("====================")


# ---- Main Loop ----
while True:
    show_menu()
    choice = input("Enter your choice: ")

    if choice == "1":
        register()
    elif choice == "2":
        login()
    elif choice == "3":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please enter 1, 2, or 3.")
```

### 87. Write a function that accepts a number and returns it’s factorial. You can not use any loop

```python
def fact(num):
    factorial = 1
    for i in range(1, num + 1):
        factorial = factorial * i
    return(factorial)
        
num = int(input("Enter a number: "))

x = fact(num)

print("Factorial =", x)
```

### 89. Write a program to find all pairs in a list whose sum equals a given target number.

```python
numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))
target = int(input("Enter target sum: "))
pairs = []
seen = set()
for num in numbers:
    complement = target - num
    if complement in seen:
        pairs.append((complement, num))
    seen.add(num)
print("Pairs:", pairs)
```

### 90. Write a program to find the second largest number in a list without using sort or max.

```python
numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))
first = second = float('-inf')
for num in numbers:
    if num > first:
        second = first
        first = num
    elif num > second and num != first:
        second = num
print("Second Largest:", second)
```

### 93. Write a program to flatten a nested list (list inside a list) into a single list without using any built-in flatten method.

```python
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

nested = [[1, 2, [3, 4]], [5, [6, [7, 8]]], 9]
print(flatten(nested))
```

### 97. Write a program to check whether two strings are anagrams of each other without using sort.


```python
s1 = input("Enter first string: ").lower().replace(" ", "")
s2 = input("Enter second string: ").lower().replace(" ", "")
freq = {}
for ch in s1:
    freq[ch] = freq.get(ch, 0) + 1
for ch in s2:
    freq[ch] = freq.get(ch, 0) - 1
print("Anagram" if all(v == 0 for v in freq.values()) and len(s1) == len(s2) else "Not an Anagram")
```

### 98. Write a program to generate a Caesar cipher: shift each letter in the input string by a given number of positions.


```python
text = input("Enter text: ")
shift = int(input("Enter shift value: "))
result = ""
for ch in text:
    if ch.isalpha():
        base = ord('A') if ch.isupper() else ord('a')
        result += chr((ord(ch) - base + shift) % 26 + base)
    else:
        result += ch
print("Encoded:", result)
```
